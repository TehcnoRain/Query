```python
# Specify the database name and collection name
database_name = "sample_supplies"
collection_name = "sales"

# Get the specified database and collection
db = client[database_name]
collection = db[collection_name]

# Aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": "$storeLocation",
            "document_count": {"$sum": 1}
        }
    },
    {
        "$match": {
            "document_count": {"$gt": 0}
        }
    },
    {
        "$lookup": {
            "from": collection_name,
            "localField": "_id",
            "foreignField": "storeLocation",
            "as": "sales"
        }
    },
    {
        "$unwind": "$sales"
    },
    {
        "$match": {
            "sales.customer.gender": "M"
        }
    },
    {
        "$group": {
            "_id": "$_id",
            "total_count": {"$sum": 1}
        }
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the result
print(result)

# Add spacing
print("\n" + "=" * 30 + "\n")

# Print the result in a nice format
for entry in result:
    print(f"Store Location: {entry['_id']}")
    print(f"Total Count (Gender M): {entry['total_count']}")
    print("-" * 30)
```

```python
# Specify the database name and collection name
database_name = "sample_supplies"
collection_name = "sales"

# Get the specified database and collection
db = client[database_name]
collection = db[collection_name]


# Aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": "$storeLocation",
            "total_count": {"$sum": 1},
            "gender_M_count": {
                "$sum": {"$cond": [{"$eq": ["$customer.gender", "M"]}, 1, 0]}
            },
            "gender_F_count": {
                "$sum": {"$cond": [{"$eq": ["$customer.gender", "F"]}, 1, 0]}
            },
            "documents": {"$push": "$$ROOT"}
        }
    },
    {
        "$match": {
            "$or": [
                {"gender_M_count": {"$gt": 0}},
                {"gender_F_count": {"$gt": 0}}
            ]
        }
    },
    {
        "$unwind": "$documents"
    },
    {
        "$project": {
            "storeLocation": "$_id",
            "total_count": "$total_count",
            "gender_M_count": "$gender_M_count",
            "gender_F_count": "$gender_F_count",
            "customer_gender": "$documents.customer.gender"
        }
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the result
for document in result:
    print(document)
```

```python
# Verifying Aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": {
                "storeLocation": "$storeLocation",
                "customer_gender": "$customer.gender"
            },
            "gender_count": {"$sum": 1},
        }
    },
    {
        "$match": {
            "$or": [
                {"_id.customer_gender": "M"},
                {"_id.customer_gender": "F"}
            ]
        }
    },
    {
        "$group": {
            "_id": {
                "storeLocation": "$_id.storeLocation",
            },
            "total_count": {"$sum": "$gender_count"},
            "gender_counts": {
                "$push": {
                    "gender": "$_id.customer_gender",
                    "count": "$gender_count"
                }
            }
        }
    }
]

# Execute the Verifying aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the result
for document in result:
    print(f"City: {document['_id']['storeLocation']}, Total Count: {document['total_count']}")
    for gender_count in document['gender_counts']:
        print(f"  Gender: {gender_count['gender']}, Count: {gender_count['count']}")
```

```python
from pymongo import MongoClient
from bson.decimal128 import Decimal128

# Specify the database name and collection name
database_name = "sample_airbnb"
collection_name = "listingsAndReviews"

# Get the specified database and collection
db = client[database_name]
collection = db[collection_name]

# Aggregation pipeline to find average price for listings with prices between 200 and 299
pipeline = [
    {
        "$match": {
            "price": {"$gte": 200, "$lte": 299}
        }
    },
    {
        "$group": {
            "_id": None,
            "average_price": {"$avg": "$price"}
        }
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the average price result
if result:
    average_price = Decimal128(result[0]["average_price"])
    average_price_float = float(average_price.to_decimal())
    print(f"The average price for listings with prices between 200 and 299 is: ${average_price_float:.2f}")
else:
    print("No matching listings found.")
```

```python
# Specify the database name and collection name
database_name = "sample_airbnb"
collection_name = "listingsAndReviews"

# Get the specified database and collection
db = client[database_name]
collection = db[collection_name]

# Aggregation pipeline to find average price for listings with prices between 200 and 299
pipeline = [
    {
        "$match": {
            "price": {"$gte": 200, "$lte": 299}
        }
    },
    {
        "$group": {
            "_id": None,
            "average_price": {"$avg": "$price"},
            "totalCount": {"$sum": 1},
        }
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the average price result
if result:
    average_price = result[0]["average_price"]
    total_count = result[0]["totalCount"]
    print(f"The average price for listings with prices between 200 and 299 is: {average_price}")
    print(f"Total Count of Documents: {total_count}")
else:
    print("No matching listings found.")
```
